### code to crop video frames - works only with AVI

```python
import cv2
```

```python
video_file = "176"
start_sec = 20
end_sec = 32
fps = 30
```

```python
# Define video path and output path
video_path = f"{video_file}.avi"
output_path = f"{video_file} - crop.avi"

# Start and end times in seconds (adjust as needed)
start_time = start_sec*fps  # (secs*30frames)
end_time = end_sec*fps  # (secs*30frames)

# Read the video using OpenCV
cap = cv2.VideoCapture(video_path)
fps = cap.get(cv2.CAP_PROP_FPS)

# Temporary video file for processing (adjust codec if needed)
temp_path = "temp.avi"
fourcc = cv2.VideoWriter_fourcc(*'XVID')
temp_writer = cv2.VideoWriter(temp_path, fourcc, fps,
                              (int(cap.get(cv2.CAP_PROP_FRAME_WIDTH)), int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))))

frame_count = 0
while True:
    ret, frame = cap.read()

    if not ret:
        break

    if start_time <= frame_count <= end_time:
        # Write frame to temporary video within the desired range
        temp_writer.write(frame)

    frame_count += 1
# print(frame_count)

cap.release()
temp_writer.release()

# Now, read the temporary video and write the desired frames to the output
cap = cv2.VideoCapture(temp_path)
writer = cv2.VideoWriter(output_path, fourcc, fps,
                         (int(cap.get(cv2.CAP_PROP_FRAME_WIDTH)), int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))))

while True:
    ret, frame = cap.read()

    if not ret:
        break

    writer.write(frame)

cap.release()
writer.release()

# Delete temporary video
import os
os.remove(temp_path)
```
